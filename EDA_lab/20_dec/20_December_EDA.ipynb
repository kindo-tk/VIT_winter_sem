## Write a program to find the maximum of 3 numbers using `if-else`

```python
a = int(input("Enter the first number:"))
b = int(input("Enter the second number:"))
c = int(input("Enter the third number:"))
if a>=b and a>=c:
    print("Largest is:",a)
elif b>=a and b>=c:
    print("Largest is:",b)
else:
    print("Largest is:",c)
```

## Print all the prime numbers in 1 to 50

```python
def is_prime(n):
    if n<2:
        return False
    for i in range(2,int(n**.5)+1):
        if n%i == 0:
            return False
    return True
for i in range(1,51):
    if is_prime(i):
        print(i," is a prime")
        
```

## Use lambda function to filterout  even number from the list

```python
og_list = [ x for x in range(1,51)]
k = lambda x: x%2==0
even_list = [x for x in og_list if k(x)]
print("Original list:\n",og_list)
print("Filtered list:\n",even_list)
```

## Alternate way

```python
og_list = [ x for x in range(1,51)]
even_list = list(filter(lambda x:x%2==0,og_list))
print("Original list:\n",og_list)
print("Filtered list:\n",even_list)
```
