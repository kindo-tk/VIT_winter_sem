# Simple Linear Regression Exercise Problems
 What is the relationship between the randomly generated 'x' values ranging from 2 to 10 and the
 corresponding 'y' values? How is this relationship represented by the linear regression model
 fitted to the data, and what are the key metrics used to evaluate the model's performance

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

# Generate random data-set
#np.random.seed(0) #to ensure that everytime we run the code same sequence of number is generated
x = np.random.uniform(2, 10, size=(100, 1))  
y = 2 + 3*x + np.random.rand(100, 1)

#Model initialization
regression_model = LinearRegression()

# Fit the data (train the model)
regression_model.fit(x, y)

# Predict
y_predicted = regression_model.predict(x)

# Model evaluation
rmse = mean_squared_error(y, y_predicted)

r2 = r2_score(y, y_predicted)

#Printing evaluation metrics
print('Root Mean Squared Error:', rmse)
print('R^2 Score:', r2)

# Coefficients
print('Coefficients:', regression_model.coef_)

# Intercepts
print('Intercept:', regression_model.intercept_)

# Plotting
plt.scatter(x, y, s=10, label='Original Data')
plt.plot(x, y_predicted, color='r', label='Fitted Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```
