# Q - A soft drink bottle is analyzing the vending machine service routes in his distribution system. He is interested in predicting the amount of time required by the route driver to service the vending machines in an outlet. this service activity includes stocking  the machine with beverage products and minor maintenance or house keeping. The industrial engineer responsible for the study has suggested that the two most important variables affecting the delivery time(y) and the number of cases of product stocked(x1) and the distance walked by the route driver (x2). The engineer has collected 25 observation on delivery time, which are shown

### We will predict Dilevery Time(y) wwith the help of Number of cases (x1) and Distance(x2):

```python
# import the necessary libraries
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
y = np.array([16.68,11.50,12.03,14.88,13.75,18.11,8.00,17.83,79.24,21.50,40.33,21.00,13.50,19.75,24.00,29.00,15.35,19.00,9.50,35.10,17.90,52.32,18.75,19.83,10.75])
y
```

```python
x1 = np.array([7,3,3,4,6,7,2,7,30,5,16,10,4,6,9,10,6,7,3,17,10,26,9,8,4])
x1
```

```python
x2 = np.array([560,220,340,80,150,330,110,210,1460,605,688,215,255,462,448,776,200,132,36,770,140,810,450,635,150])
x2
```

```python
## converting the data into a dataframe
df = pd.DataFrame({"Delivery Time(y)":y,"Number of Cases (x1)":x1,"Distance(x2)":x2})
df
```

```python
sns.pairplot(df,diag_kind=None)
plt.suptitle("Scatterplot matrix of delivery Time data")
plt.show()
```

## 3D scatter plot

```python
plt.figure(figsize=(6,6))
ax = plt.axes(projection = '3d')
ax.scatter(x1,x2,y,c='blue',marker = 'o')
ax.set_xlabel("Number of cases")
ax.set_ylabel("Distance (ft)")
ax.set_zlabel("Delivery time (min)")
ax.set_title("3D scatter plot of Delivery Time vs Number of cases and distance")
plt.show()
```

## 3D scatter plot enhanced

```python
plt.figure(figsize=(10,8))
ax = plt.axes(projection = '3d')
sc = ax.scatter(x1,x2,y,c=y,cmap = 'viridis',s = 50 ,alpha = 0.9) # s is for size of the scatter plot marker
ax.set_xlabel("Number of cases")
ax.set_ylabel("Distance (ft)")
ax.set_zlabel("Delivery time (min)")
ax.set_title("3D scatter plot of Delivery Time data")
cbar = plt.colorbar(sc,ax = ax, pad= 0.1) # pad controls the space between the plot and color bar
cbar.set_label("Delivery Time(y)",rotation = 270,labelpad = 15)
plt.show()
```

## OLS estimation:

## Compute the coefficients (β0,β1,β2) of a multiple linear regression model using matrix operations

```python
n = len(y)
x = np.column_stack((np.ones(n),x1,x2))
x
```

```python
xtx = np.dot(x.T,x)
xtx
```

```python
xtx_inv = np.linalg.inv(xtx)
xtx_inv
```

```python
xty = np.dot(x.T,y)
xty
```

```python
beta_hat = np.dot(xtx_inv,xty)

print("OLS Estimators (β0,β1,β2):")
print(beta_hat)
```

### Regression plane visualization

```python
x1_grid, x2_grid = np.meshgrid(np.linspace(min(x1),max(x1),20),np.linspace(min(x2),max(x2),20))
beta0,beta1,beta2 = beta_hat

y_pred = beta0+beta1*x1_grid+beta2*x2_grid

plt.figure(figsize=(10,8))
ax = plt.axes(projection = '3d')
sc = ax.scatter(x1,x2,y,c=y,cmap = 'viridis',s=50,alpha = 0.9,label = "Actual data")

ax.plot_surface(x1_grid,x2_grid,y_pred,color = 'orange',alpha = 0.5,rstride = 100, cstride = 100 )
ax.set_xlabel("Number of cases (x1)")
ax.set_ylabel("Distance (ft) (x2)")
ax.set_zlabel("Delivery time (min) (y)")
ax.set_title("3D scatter plot with regression plane")

cbar = plt.colorbar(sc,ax=ax,pad = 0.1)
cbar.set_label("Delivery Time (y)",rotation = 270, labelpad = 15)
plt.legend()
plt.show()
```
