## Heights of Fathers and Daughters

The following data represents the heights of fathers and their daughters:

| Height of Father (cm) | Height of Daughter (cm) |
|------------------------|-------------------------|
| 65                     | 67                      |
| 66                     | 68                      |
| 67                     | 64                      |
| 67                     | 69                      |
| 68                     | 72                      |
| 69                     | 70                      |
| 71                     | 69                      |
| 73                     | 73                      |

### Tasks:
1. Fit a linear regression model to predict the height of the daughter (\(y\)) based on the height of the father (\(x\)).
2. Use the fitted model to predict the height of a daughter when the fatherâ€™s height is 70 cm.
3. Compute the residuals for the given data points.
4. Plot the fitted regression model and the given data points on a graph.


```python
# importing the required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd

# Data
father_heights = np.array([65, 66, 67, 67, 68, 69, 71, 73])
daughter_heights = np.array([67, 68, 64, 69, 72, 70, 69, 73])

# Fit the linear regression model
model = LinearRegression()
model.fit(father_heights.reshape(-1, 1), daughter_heights)

# Get coefficients
slope = model.coef_[0]
intercept = model.intercept_

# Predict daughter's heights for all father's heights
predicted_daughter_heights = model.predict(father_heights.reshape(-1, 1))

# Predict the daughter's height for a specific father's height 
father_height_70 = np.array([[70]])
predicted_height_70 = model.predict(father_height_70)

# Calculate residuals
residuals = daughter_heights - predicted_daughter_heights

# Create the table with x, y, y_pred, xy, x^2
data = {
    "x (Father's Height)": father_heights,
    "y (Daughter's Height)": daughter_heights,
    "y_pred (Predicted y)": predicted_daughter_heights,
    "xy": father_heights * daughter_heights,
    "x^2": father_heights**2,
}

# Summations
summations = {
    "x (Father's Height)": np.sum(father_heights),
    "y (Daughter's Height)": np.sum(daughter_heights),
    "y_pred (Predicted y)": np.sum(predicted_daughter_heights),
    "xy": np.sum(father_heights * daughter_heights),
    "x^2": np.sum(father_heights**2),
}

# Convert data to DataFrame
df = pd.DataFrame(data)
df.loc["Sum"] = df.sum()

# Display the table
print(df)

# Plot the regression line and data points
plt.scatter(father_heights, daughter_heights, color='blue', label='Data Points')
plt.plot(father_heights, predicted_daughter_heights, color='red', label='Regression Line')
plt.scatter(70, predicted_height_70, color='green', label='Prediction (70 cm Father)')

plt.xlabel("Height of Father (cm)")
plt.ylabel("Height of Daughter (cm)")
plt.title("Linear Regression: Father-Daughter Heights")
plt.legend()
plt.grid()
plt.show()

# Print regression equation and predictions
print(f"Regression Equation: y = {slope:.4f}x + {intercept:.4f}")
print("Predicted Heights for Given x (Father's Heights):", predicted_daughter_heights)
print(f"Predicted Daughter's Height for 70 cm Father: {predicted_height_70[0]:.2f} cm")
print("Residuals:", residuals)
```
