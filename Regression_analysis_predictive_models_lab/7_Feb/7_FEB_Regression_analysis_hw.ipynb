## Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm 
```

## Reading the dataset

```python
df = pd.read_excel(r"D:\study material\VIT_Data_Science\Winter_Sem\Regression Analysis and Predictive Models Lab\7_feb\advertising.xlsx")
df
```

```python
df.info()
```

## Perform Linear Regression using sklearn:
- Define TV as the independent variable (X) and Sales as the dependent variable
(y).


```python
x_tv = df['TV'].values.reshape(-1,1)
y = df['Sales'].values

model_tv = LinearRegression()
model_tv.fit(x_tv,y)
y_pred_tv = model_tv.predict(x_tv)
residuals_tv = y - y_pred_tv
```

### Scatter plot of TV vs Sales along with the regression line and the residuals


```python
## creating plot to visualize the regression line

plt.figure(figsize = (10,6))
plt.scatter(x = x_tv, y = y, color = 'blue', label = 'Data points', alpha = 0.5)
plt.plot(x_tv,y_pred_tv, color = 'red', label = "Regression Line")

## Adding lines showing the residuals(the vertical distance between actual and predicted value)
for i in range(len(x_tv)):
    plt.plot([x_tv[i],x_tv[i]],[y[i],y_pred_tv[i]], color = 'grey', lw=1)

plt.title("Least square fit for sales vs TV advertising")
plt.xlabel("TV advertising budget(in thousands of dollars)")
plt.ylabel("Sales (in thousand of units)")
plt.legend()
plt.show()
```

## Perform Linear Regression using statsmodels:


```python
import statsmodels.api as sm 
x_tv = df['TV']
y = df['Sales']

# adds a constant to the independent variable 
x_with_const = sm.add_constant(x_tv) # adds a column of ones to x for the intercept form 

# Fit the OLS regression model using stats model
model_sm_tv = sm.OLS(y,x_with_const).fit()

print(model_sm_tv.summary())
# get the 95% confidence interval for the model coefficients (β₀ and β₁)

confidence_intervals_tv = model_sm_tv.conf_int(alpha = 0.05)

# print the confidence intervals for the intercept  and coefficient

print("95% confidence interval for β₀:\n",confidence_intervals_tv.iloc[0])
print("95% confidence interval for β1:\n",confidence_intervals_tv.iloc[1])

standard_error_tv = model_sm_tv.bse

print("Standard error for β₀(Intercept):",standard_error_tv.iloc[0])
print("Standard error for β1:",standard_error_tv.iloc[1])
```
