## Importing the necessary libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

## Loading the given data in a dataframe

```python
data = pd.DataFrame({'Area':[2600,3000,3200,3600,4000],'Price':[550000,565000,610000,680000,725000]})
data
```

## Visualizing the data

```python
plt.scatter(data['Area'],data['Price'],color = 'blue')
plt.xlabel('Area (Square Feet)')
plt.ylabel('Price')
plt.title("Scatter plot of area vs price")
plt.show()
```

- **Interpretation:** The scatter plot shows a positive relationship between area and price,
 indicating that as the area increases, the price generally increases.

## Preparing the data for modelling

```python
x = data['Area'].values.reshape(-1,1)
y = data['Price'].values
```

## Splitting the data into training and testing

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

## Building and training the linear regression model

```python
model = LinearRegression()
model.fit(x_train,y_train)
print("Intercept:",model.intercept_)
print("Slope(Coefficient):",model.coef_[0])
```

## Making predictions

```python
y_pred = model.predict(x_test)
y_pred
```

## Evaluating the model

```python
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print("Mean squared error is:",mse)
print("R-Squared error:",r2)
```

## Visualizing the regression line

```python
plt.scatter(data['Area'],data['Price'],color = 'blue')
plt.plot(x,model.predict(x),color = 'red',label = 'Regression line')
plt.xlabel('Area (Square Feet)')
plt.ylabel('Price')
plt.title("Simple Linear Regression: area vs price")
plt.legend()
plt.show()
```
