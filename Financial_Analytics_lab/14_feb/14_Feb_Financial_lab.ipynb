## Q1- Moving average model

xt = 0.5*xt-1 + 0.3*et-1 + et


|t|et|xt|
|-|-|-|
1|0.1|1
2|-0.5|?
3|0.9|?
4|-0.8|?
5|0.7|?

```python
import pandas as pd
import numpy as np
```

```python
t = [1,2,3,4,5]
Et = [0.1,-.5,.9,-.8,.7]
xt = [1,0,0,0,0]
for i in range(1,len(t)):
    xt[i] = 0.5*xt[i-1]+0.3*Et[i-1]+Et[i]
print(xt)
```

```python
time_series_data = {"t":t,"Et":Et,"xt":xt}
df = pd.DataFrame(data = time_series_data)
```

## Q2- Generate a random set of data Et(random noise) chose a first value of Xt and then predict the rest values

```python
## Generating random noise(uniform distribution)

et = np.random.uniform(-1,1,1000)
```

```python
et
```

```python
t = np.arange(1,1001)
t
```

```python
xt = np.zeros(1000,)
xt[0] = 3
```

```python
for i in range(1,len(t)):
    xt[i] = 0.5*xt[i-1]+0.3*et[i-1]+et[i]
print(xt)
```

```python
time_series_data_large = {"t":t,"Et":et,"xt":xt}
df1 = pd.DataFrame(data = time_series_data_large)
```

```python
df1
```

## Plotting xt vs t

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.plot(df1['t'],df1['xt'],color='red')
plt.title("Time V/s Xt values")
plt.xlabel("Time")
plt.ylabel("Xt values")
plt.show()
```

## Q3- Use et as the standard normal distribution

```python
## Generating random noise(normal distribution)

et = np.random.standard_normal(1000)
t = np.arange(1,1001)
xt = [3]
for i in range(1,len(t)):
    xt.append(0.5*xt[i-1]+0.3*et[i-1]+et[i])
time_series_data_normal_error = {"t":t,"Et":et,"xt":xt}
df2 = pd.DataFrame(data = time_series_data_normal_error)
df2
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.plot(df2['t'],df2['xt'],color='green')
plt.title("Time V/s Xt values")
plt.xlabel("Time")
plt.ylabel("Xt values")
plt.show()
```
